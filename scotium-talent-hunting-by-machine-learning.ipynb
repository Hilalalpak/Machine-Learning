```python
def grab_col_names(dataframe, cat_th=10, car_th=30):
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "object" or dataframe[col].dtypes.name == "category"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "object" and dataframe[col].dtypes.name != "category"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   (dataframe[col].dtypes == "object" or dataframe[col].dtypes.name == "category")]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "object" and dataframe[col].dtypes.name != "category"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    # Create a DataFrame to display results
    results = pd.DataFrame({
        'Variable Type': ['Observations', 'Variables', 'cat_cols', 'num_cols', 'cat_but_car', 'num_but_cat'],
        'Count': [dataframe.shape[0], dataframe.shape[1], len(cat_cols), len(num_cols), len(cat_but_car), len(num_but_cat)]
    })

    return results, cat_cols, num_cols, num_but_cat, cat_but_car


results, cat_cols, num_cols, num_but_cat, cat_but_car = grab_col_names(df)

print("Results:")
print(results)
print("-------------------------------")
print("Numeric but Categoric:")
print(num_but_cat)
```

```python
# Optimal number of clusters using the Elbow method
inertia = []
ssd = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=52)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    ssd.append(kmeans.inertia_)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), ssd, marker='o', linestyle='-', color='r', markerfacecolor='b')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.title('K-Means Performance Metrics with SSE/SSR/SSD')
plt.axvline(x=2, color='r', linestyle='--')
plt.xticks(range(1, 11))
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show();
```

```python
kmeans = KMeans(n_clusters=2, random_state=52, n_init=10, max_iter=300).fit(X)
params = kmeans.get_params()
```

```python
def base_models(X, y, rs, scoring):
    print(f"Evaluating Base Models with {scoring} scoring...")
    classifiers = [('LR', LogisticRegression(random_state=rs)),
                   ('KNN', KNeighborsClassifier()),
                   ("CART", DecisionTreeClassifier(random_state=rs)),
                   ("RF", RandomForestClassifier(random_state=rs)),
                   ('Adaboost', AdaBoostClassifier(random_state=rs)),
                   ('GBM', GradientBoostingClassifier(random_state=rs)),
                   ('XGBoost', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=rs)),
                   ('CatBoost', CatBoostClassifier(verbose=False, random_state=rs))]

    results = []
    for name, classifier in classifiers:
        cv_results = cross_validate(classifier, X, y, cv=3, scoring=scoring)
        results.append([name, round(cv_results['test_score'].mean(), 4)])

    results_df = pd.DataFrame(results, columns=['Model', scoring.capitalize()])
    results_df.sort_values(by=scoring.capitalize(), ascending=False, inplace=True)
    return results_df

# Assisted func.
def colored_print(value, top_values, score):
    if value in top_values:
        return f"{fore(9)}{value}{style(0)}"
    elif value == score:
        return f"{fore(1)}{value}{style(0)}"
    else:
        return f"{value}"

# Scores
scores = ["roc_auc", "f1", "precision", "recall"]
tables = []

for score in scores:
    result_df = base_models(X, y, 52, score)
    top_values = result_df.iloc[:3]["Model"].tolist()
    result_df["Model"] = result_df.apply(lambda row: colored_print(row["Model"], top_values, score), axis=1)
    tables.append(result_df)

# Show tables
for table, score in zip(tables, scores):
    print(f"\nPerformance for {score} scoring (sorted):\n")
    print(tabulate(table, headers='keys', tablefmt='pretty', showindex=False))
```
