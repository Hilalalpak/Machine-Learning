```python
# Convert to meaningful categories

def standardize_categories(df_original, df_train):
    valid_categories = {col: set(df_original[col].unique()) 
                        for col in df_original.columns 
                        if df_original[col].dtype == 'object'}
    
    for col in valid_categories.keys():
        if col in df_train.columns:
            mask = ~df_train[col].isin(valid_categories[col])
            df_train.loc[mask, col] = 'Unknown'
    
    return df_train

df_train = standardize_categories(df_original, df_train)
df_test = standardize_categories(df_original, df_test)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots


def plot_top_value_counts_interactive(dataframe, categorical_columns):
    num_cols = 3 
    num_rows = (len(categorical_columns) + num_cols - 1) // num_cols  

    fig = make_subplots(rows=num_rows, cols=num_cols, subplot_titles=categorical_columns, 
                        vertical_spacing=0.06, horizontal_spacing=0.1)

    for i, column in enumerate(categorical_columns):
        # Get the top 10 value counts
        top_values = dataframe[column].value_counts().head(10)
        values = top_values.values
        labels = top_values.index
        total = values.sum()

        # Calculate percentages
        percentages = (values / total) * 100

        # Create bar trace
        bar_trace = go.Bar(
            x=labels,
            y=values,
            text=[f'{v:.1f} ({p:.1f}%)' for v, p in zip(values, percentages)],
            textposition='none',
            marker=dict(color='skyblue')
        )
        
        # Determine row and column index for subplot
        row = i // num_cols + 1
        col = i % num_cols + 1
        
        fig.add_trace(bar_trace, row=row, col=col)

    # Update layout for better visualization
    fig.update_layout(
        title_text='Top 10 Value Counts for Categorical Columns',
        showlegend=False,
        height=num_rows * 300,
        width=num_cols * 500,
        autosize=False
    )

    fig.show()
        
        
        
plot_top_value_counts_interactive(df_train, cat_cols)
```

```python
from matplotlib.cm import viridis

def plot_missing_values(df, threshold=2):
    missing_values = df.isnull().sum()
    missing_values = missing_values[missing_values > threshold].sort_values(ascending=False)
    missing_percentage = (missing_values / len(df)) * 100
    
    # Use viridis colormap
    colors = plt.cm.viridis(np.linspace(0, 1, len(missing_values)))
    
    plt.figure(figsize=(10, 8))
    ax = sns.barplot(x=missing_percentage.index, y=missing_percentage.values, palette=colors)
    
    plt.title('Percentage of Missing Values Across Variables', fontsize=16)
    plt.xlabel('Variables', fontsize=14)
    plt.ylabel('Percentage of Missing Values', fontsize=14)
    plt.xticks(rotation=90, ha='center', fontsize=8)
    plt.yticks(fontsize=8)
    
    ax.grid(axis='y', linestyle='-', alpha=0.4)
    
    for i, p in enumerate(ax.patches):
        height = p.get_height()
        ax.annotate(f'{height:.1f}%', 
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=10, color='black',
                    xytext=(0, 5), textcoords='offset points')
    
    plt.tight_layout()
    plt.show()

plot_missing_values(df_train, threshold=2)
```

```python
cat_cols = [col for col in df_train.columns if df_train[col].dtype == "object" and col != "class"]

for col in cat_cols:
    df_train[col] = df_train[col].fillna('Unknown')
    df_test[col] = df_test[col].fillna('Unknown')
    
    
    
numeric_columns = ['cap-diameter', 'stem-height', 'stem-width']
numeric_imputer = SimpleImputer(strategy='median')

df_train[numeric_columns] = numeric_imputer.fit_transform(df_train[numeric_columns])
df_test[numeric_columns] = numeric_imputer.fit_transform(df_test[numeric_columns])
```

```python
def clean_unknown(df, mode_dict=None):
    for col in df.columns:
        if df[col].dtype == "object":
            if mode_dict is None:
                mode = df[col].mode()[0]
                df[col] = df[col].replace('Unknown', mode)
            else:
                df[col] = df[col].replace('Unknown', mode_dict[col])
    return df


mode_dict = {col: df_train[col].mode()[0] for col in df_train.columns if df_train[col].dtype == "object"}
df_train = clean_unknown(df_train)

df_test = clean_unknown(df_test, mode_dict)
```

```python
def plot_histograms(df, num_cols):
    plt.figure(figsize=(15, 5)) 
    
    color_palette = sns.color_palette("viridis", len(num_cols))

    for i, column in enumerate(num_cols):
        plt.subplot(1, len(num_cols), i + 1)  
        sns.histplot(data=df, x=column, kde=True, bins=20, color=color_palette[i])
        plt.title(f'Distribution of {column}', fontsize=14, fontweight='bold')
        plt.xlabel(column, fontsize=12)
        plt.ylabel('Frequency', fontsize=12)
        sns.despine()
        plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout() 
    plt.show()
    
    
plot_histograms(df_train, num_cols)    
```

```python
# Function to calculate lower and upper thresholds
def outlier_thresholds(dataframe, col_name, q1=0.10, q3=0.90):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

# Function to check for outliers in a specific column
def check_outlier(dataframe, col_name):
    if pd.api.types.is_numeric_dtype(dataframe[col_name]):
        low_limit, up_limit = outlier_thresholds(dataframe, col_name)
        return dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)][col_name]
    else:
        return pd.Series([])

# Function to replace outliers with defined thresholds
def replace_with_thresholds(dataframe, variable, q1=0.10, q3=0.90):
    low_limit, up_limit = outlier_thresholds(dataframe, variable, q1, q3)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit

# Selecting all numeric columns
numeric_vars = df_train.select_dtypes(include=['float64']).columns.tolist()

# Iterating through each numeric column to check and handle outliers
for col in numeric_vars:
    outliers = check_outlier(df_train, col)
    if not outliers.empty:
        print(f"Outliers found in {col}. Handling outliers...")
        replace_with_thresholds(df_train, col)
```

```python
############
# Encoding
############

def encode_categories(df_train, df_test):
    ohe_cols = ['does-bruise-or-bleed', 'has-ring', 'gill-spacing', 'season']
    
    for col in ohe_cols:
        dummies_train = pd.get_dummies(df_train[col], prefix=col, drop_first=True)
        dummies_test = pd.get_dummies(df_test[col], prefix=col, drop_first=True)
        
        df_train = pd.concat([df_train, dummies_train], axis=1)
        df_test = pd.concat([df_test, dummies_test], axis=1)
        
        df_train.drop(columns=[col], inplace=True)
        df_test.drop(columns=[col], inplace=True)
    
    le_cols = ['cap-shape', 'cap-surface', 'gill-attachment', 'stem-surface', 'ring-type', 'habitat', 'cap-color', 'gill-color', 'stem-color']
    
    for col in le_cols:
        le = LabelEncoder()
        df_train[col] = le.fit_transform(df_train[col])
        df_test[col] = le.transform(df_test[col])
    
    return df_train, df_test



df_train, df_test = encode_categories(df_train, df_test)


df_train['class'] = df_train["class"].map({'e': 0, 'p': 1})
```

```python
############
# Standardization
############


def standardize_data(df_train, df_test):
    numeric_columns = df_train.select_dtypes(include=[np.number]).columns
    numeric_columns = [col for col in df_train.columns if col != "class"]
    scaler = StandardScaler()

    scaler.fit(df_train[numeric_columns])

    df_train_scaled = df_train.copy()
    df_train_scaled[numeric_columns] = scaler.transform(df_train[numeric_columns])

    df_test_scaled = df_test.copy()
    df_test_scaled[numeric_columns] = scaler.transform(df_test[numeric_columns])

    return df_train_scaled, df_test_scaled, scaler


df_train, df_test, scaler = standardize_data(df_train, df_test)
```

```python
#The best parameters were determined using RandomizedSearchCV, 
#but due to the data size, the process was very time-consuming. 
#Therefore, the code was removed and the optimal parameters were used directly.

model = XGBClassifier(                    
    alpha=0.1,                   
    subsample=0.8,     
    colsample_bytree=0.6,  
    objective='binary:logistic',
    max_depth=14,             
    min_child_weight=7,         
    gamma=1e-6,                
    random_state=42,                 
    n_estimators=100
    )


def mcc_metric(y_pred, dmatrix):
    y_true = dmatrix.get_label()
    y_pred = (y_pred > 0.5).astype(int) 
    mcc = matthews_corrcoef(y_true, y_pred)
    return 'mcc', mcc



XGB = model.fit(
    train_X, 
    train_y, 
    eval_set=[(test_X, test_y)],
    eval_metric=mcc_metric,
    verbose=0)
```
