```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "object" or dataframe[col].dtypes.name == "category"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "object" and dataframe[col].dtypes.name != "category"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   (dataframe[col].dtypes == "object" or dataframe[col].dtypes.name == "category")]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "object" and dataframe[col].dtypes.name != "category"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols

cat_cols, num_cols = grab_col_names(df)
cat_cols
num_cols
```

```python
# Calculates the lower and upper thresholds to identify outliers for a given column in the dataframe

def outlier_thresholds(dataframe, col_name, q1=0.10, q3=0.90):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit


# Checks if a specific column in the dataframe contains any outliers based on calculated thresholds

def check_outlier(dataframe, col_name):
    if pd.api.types.is_numeric_dtype(dataframe[col_name]):
        low_limit, up_limit = outlier_thresholds(dataframe, col_name)
        return (dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)
    else:
        return pd.Series(False, index=dataframe.index)


# Checks all columns in the dataframe for the presence of outliers and returns a dictionary with the results        

def check_all_columns_outliers(dataframe):
    results = {}
    for col in dataframe.columns:
        results[col] = check_outlier(dataframe, col).any() if pd.api.types.is_numeric_dtype(dataframe[col]) else False
    return results
```

```python
## The function perform_of_models evaluates the performance of various machine learning models using k-fold cross-validation 

def perform_of_models(X, y):
    classifiers = {
        "LR": LogisticRegression(),
        "KNN": KNeighborsClassifier(),
        "RFC": RandomForestClassifier(),
        "CBC": CatBoostClassifier(silent=True),
        "XGB": XGBClassifier(),
        "LGBM": LGBMClassifier(),
        "DT": DecisionTreeClassifier(),
        "GBC": GradientBoostingClassifier(),
        "ABC": AdaBoostClassifier()
    }

    def evaluate_model(y_true, y_pred):
        all_metrics = {
            "Accuracy": accuracy_score(y_true, y_pred),
            "Precision": precision_score(y_true, y_pred),
            "Recall": recall_score(y_true, y_pred),
            "F1": f1_score(y_true, y_pred),
            "AUC": roc_auc_score(y_true, y_pred)
        }
        return all_metrics

    model_performances = []

    for label, model in tqdm(classifiers.items()):
        n_splits = 3  
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=47)
        performances = np.zeros((n_splits, 5))  
        
        X_values = X.values
        i = 0
        for train_idx, test_idx in kf.split(X_values, y):
            train_set = X_values[train_idx], y.iloc[train_idx]
            test_set = X_values[test_idx], y.iloc[test_idx]
            model.fit(*train_set)
            y_pred = model.predict(test_set[0])
            perf = evaluate_model(test_set[1], y_pred)
            performances[i, :] = list(perf.values())
            i += 1
        
        model_performances.append(
            pd.Series(np.mean(performances, axis=0), index=list(perf.keys()), name=label)
        )

    performances_df = pd.concat(model_performances, axis=1)

    avg_f1 = performances_df.loc["F1"].mean()

    # Creating the figure for the graph using plotly
    fig = go.Figure()

    # Pastel renkler
    colors = ["#7b68ee", "#98fb98", "#87ceeb", "#ffa500", "#ff6347", "#ffc0cb", "#ff69b4", "#00ced1", "#dda0dd"]

    # Adding traces for each metric in the specified order
    for metric, color in zip(["Accuracy", "AUC", "Recall", "Precision", "F1"], colors):
        fig.add_trace(go.Bar(
            x=performances_df.columns,
            y=performances_df.loc[metric],
            marker_color=color,
            name=metric,
            text=[f"{val:.3f}" for val in performances_df.loc[metric]],
            textposition='auto'
        ))

    # Updating the layout of the figure
    fig.update_layout(
        xaxis_title="Models",
        yaxis_title="Metric Score",
        title="Comparison of Metrics for Different Models",
        barmode='group',
        bargap=0.15,
        bargroupgap=0.1
    )

    # Displaying the graph
    fig.show()
    
    
perform_of_models(X, y)
```

```python
def evaluate_model_plotly(model, train_features, train_target, test_features, test_target):
    eval_stats = {}
    
    # Create subplots with shared y axis
    fig = sp.make_subplots(rows=1, cols=3, subplot_titles=("F1 Score", "ROC Curve", "Precision-Recall Curve"),
                           horizontal_spacing=0.1, vertical_spacing=0.1)
    
    for type, features, target, ax_idx in (('train', train_features, train_target, 1), ('test', test_features, test_target, 2)):
        
        eval_stats[type] = {}
    
        pred_target = model.predict(features)
        pred_proba = model.predict_proba(features)[:, 1]
        
        # F1 Score
        f1_thresholds = np.arange(0, 1.01, 0.05)
        f1_scores = [metrics.f1_score(target, pred_proba>=threshold) for threshold in f1_thresholds]
        
        max_f1_score_idx = np.argmax(f1_scores)
        fig.add_trace(go.Scatter(x=f1_thresholds, y=f1_scores, mode='lines', name=f'{type}, max={f1_scores[max_f1_score_idx]:.2f} @ {f1_thresholds[max_f1_score_idx]:.2f}', 
                                 line=dict(color='blue' if type == 'train' else 'green')), row=1, col=1)
        # setting crosses for some thresholds
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(f1_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            fig.add_trace(go.Scatter(x=[f1_thresholds[closest_value_idx]], y=[f1_scores[closest_value_idx]], mode='markers', marker=dict(color=marker_color, symbol='x', size=10),
                                     showlegend=False), row=1, col=1)
        fig.update_xaxes(title_text='Threshold', range=[-0.02, 1.02], row=1, col=1)
        fig.update_yaxes(title_text='F1', range=[-0.02, 1.02], row=1, col=1)

        # ROC Curve
        fpr, tpr, roc_thresholds = metrics.roc_curve(target, pred_proba)
        roc_auc = metrics.roc_auc_score(target, pred_proba)
        fig.add_trace(go.Scatter(x=fpr, y=tpr, mode='lines', name=f'{type}, ROC AUC={roc_auc:.2f}', line=dict(color='blue' if type == 'train' else 'green')), row=1, col=2)
        # setting crosses for some thresholds
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(roc_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            fig.add_trace(go.Scatter(x=[fpr[closest_value_idx]], y=[tpr[closest_value_idx]], mode='markers', marker=dict(color=marker_color, symbol='x', size=10),
                                     showlegend=False), row=1, col=2)
        fig.add_shape(type='line', line=dict(dash='dash'), x0=0, x1=1, y0=0, y1=1, row=1, col=2)
        fig.update_xaxes(title_text='FPR', range=[-0.02, 1.02], row=1, col=2)
        fig.update_yaxes(title_text='TPR', range=[-0.02, 1.02], row=1, col=2)

        # Precision-Recall Curve
        precision, recall, pr_thresholds = metrics.precision_recall_curve(target, pred_proba)
        aps = metrics.average_precision_score(target, pred_proba)
        fig.add_trace(go.Scatter(x=recall, y=precision, mode='lines', name=f'{type}, AP={aps:.2f}', line=dict(color='blue' if type == 'train' else 'green')), row=1, col=3)
        # setting crosses for some thresholds
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(pr_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            fig.add_trace(go.Scatter(x=[recall[closest_value_idx]], y=[precision[closest_value_idx]], mode='markers', marker=dict(color=marker_color, symbol='x', size=10),
                                     showlegend=False), row=1, col=3)
        fig.update_xaxes(title_text='Recall', range=[-0.02, 1.02], row=1, col=3)
        fig.update_yaxes(title_text='Precision', range=[-0.02, 1.02], row=1, col=3)

        eval_stats[type]['Accuracy'] = metrics.accuracy_score(target, pred_target)
        eval_stats[type]['F1'] = metrics.f1_score(target, pred_target)

    # Update layout
    fig.update_layout(
        height=500,
        width=1200,
        title_text="Model Evaluation Metrics",
        showlegend=True,
        legend=dict(x=8, y=0.98, bgcolor='rgba(255, 255, 255, 0.5)'),
        font=dict(size=12),
        margin=dict(l=50, r=50, t=100, b=50),
    )

    # Display the figure
    fig.show()


evaluate_model_plotly(best_log_model, X_train, y_train, X_test, y_test)
```
